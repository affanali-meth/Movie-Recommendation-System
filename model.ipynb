```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df = df[['title', 'overview', 'genres', 'tagline', 'vote_average', 'popularity']]
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['title'])
df['overview'] = df['overview'].fillna('')
import ast
df['genres'] = df['genres'].apply(lambda x: " ".join([i['name'] for i in ast.literal_eval(x)]))
df['tagline'] = df['tagline'].fillna('')
```

```python
df['tags'] = df['overview'] + " " + df['genres'] + " " + df['tagline']
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # Convert to string (safety)
    text = str(text).lower()

    # Remove punctuation + numbers
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Split into words
    words = text.split()

    # Remove stopwords
    words = [word for word in words if word not in stop_words]

    # Lemmatize
    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df = df.reset_index(drop=True)
```

```python
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(
    max_features=50000,
    ngram_range=(1, 2),
    stop_words='english'
)
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
# similarity = linear_kernel(tfidf_matrix, tfidf_matrix) this will kill

from sklearn.metrics.pairwise import linear_kernel

from sklearn.metrics.pairwise import linear_kernel

def recommend(title, n=10):
    if title not in indices:
        return ["Movie not found"]

    idx = indices[title]

    # Compute similarity only for this movie → SAFE!
    sim_scores = linear_kernel(tfidf_matrix[idx], tfidf_matrix).flatten()

    # Sort by score
    similar_indices = sim_scores.argsort()[::-1][1:n+1]

    return df['title'].iloc[similar_indices].tolist()
```

```python
recommend("Comic Book Confidential", 10)
```
